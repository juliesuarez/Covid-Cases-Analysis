```python
#Data Loading & Exploration

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#This function reads the data from your CSV file and loads it into a DataFrame
df = pd.read_csv('owid-covid-data.csv')


#Shows data's structure.
print("First 5 rows of the dataset:")
display(df.head())

#Check all the column names
print("\nColumn names:")
print(df.columns)

#tells you the data type of each column
print("\nDataset Info:")
df.info()

#Identify the count of missing values in each column
print("\nMissing values per column:")
print(df.isnull().sum())

#Data Cleaning

# Convert the 'date' column to a proper datetime format
df['date'] = pd.to_datetime(df['date'])

#Filter the data for a few countries of interest to make analysis manageable
countries_of_interest = ['Uganda','Kenya', 'United States', 'India', 'United Kingdom', 'Brazil']
df_filtered = df[df['location'].isin(countries_of_interest)].copy()

#Handle missing numeric values
# For columns like new cases or new deaths, a missing value often means zero were reported.
#fill these with 0.
cols_to_fill_zero = ['new_cases', 'new_deaths', 'total_vaccinations']
for col in cols_to_fill_zero:
    df_filtered[col] = df_filtered[col].fillna(0)

# Set display options to see the full output-large amounts of data make the data slow
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

# Display the first few rows of our new, cleaned DataFrame
print("Cleaned and filtered data for selected countries:")
display(df_filtered.head())

# Visualizations

# Set a nice style for the plots
sns.set_style('whitegrid')
plt.figure(figsize=(14, 8))

# Total Cases Over Time
plt.subplot(2, 1, 1) # Create a subplot for the first graph
sns.lineplot(data=df_filtered, x='date', y='total_cases', hue='location')
plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases (in millions)')

#Total Deaths Over Time
plt.subplot(2, 1, 2) # Create a subplot for the second graph
sns.lineplot(data=df_filtered, x='date', y='total_deaths', hue='location')
plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')

plt.tight_layout() # Adjust layout to prevent titles from overlapping
plt.show()

# Top 10 Countries by Total Cases (Bar Chart)
# For this, we need the latest data for all countries from the original dataframe
latest_data = df.loc[df.groupby('location')['date'].idxmax()]
top_10_cases = latest_data.nlargest(10, 'total_cases')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_cases, x='total_cases', y='location')
plt.title('Top 10 Countries by Total Confirmed Cases')
plt.xlabel('Total Cases (in hundred millions)')
plt.ylabel('Country')
plt.show()

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_filtered, x='date', y='total_vaccinations', hue='location')
plt.title('Total Vaccination Doses Administered Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations (in billions)')
plt.show()

#Build a Choropleth Map
import plotly.express as px
latest_data['total_cases_per_million'] = latest_data['total_cases_per_million'].fillna(0)

fig = px.choropleth(latest_data,
                    locations="iso_code",
                    color="total_cases_per_million",
                    hover_name="location",
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title="Global COVID-19 Cases per Million People")

fig.show()
```
